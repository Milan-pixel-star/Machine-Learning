## (2) Ordinary encoder --> it is applied o multiple column

```python
df=pd.read_csv("F:\\New folder\\ML\\CSV files\\covid_toy.csv")
df.head(2)
```

```python
df["city"].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = df.drop(['age','fever'],axis=1)
```

```python
df['city'].value_counts()
```

```python
df['cough'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Male','Female'],['Mild','Strong'],['Kolkata','Bangalore','Delhi','Mumbai'],['Yes','No']])
```

```python
oe
```

```python
oe.fit(df)
```

```python
df_new=oe.transform(df)
```

```python
oe.categories_
```

```python
df=pd.DataFrame(df_new,columns=df.columns)
```
